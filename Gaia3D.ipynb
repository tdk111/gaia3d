```python
hdul=fits.open('/Users/tom.kitching/Downloads/TGAS_Dlt3kpc.fits') #change this to the download path 
```

```python
hdul[1].header #show all the information in the table (there is lots of info, we will only use x,y,z)
```

```python
x=hdul[1].data.field(68) #starts at 0 so minus one from the above
```

```python
y=hdul[1].data.field(69) #starts at 0 so minus one from the above
```

```python
z=hdul[1].data.field(70) #starts at 0 so minus one from the above
```

```python
colour=hdul[1].data.field(71)  
```

```python
x.size #number of objects
```

```python
#make a 3d scatter plot for 100,000 objects labelled with colour
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

n = 100

# For each set of style and range settings, plot n random points in the box
# defined by x in [23, 32], y in [0, 100], z in [zlow, zhigh].
xs = x[1:100000]
ys = y[1:100000]
zs = z[1:100000]
c=colour[1:100000]
marker='o'
ax.scatter(xs, ys, zs, s=1,c=c, marker=m)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```
